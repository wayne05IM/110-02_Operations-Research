```python
# create a dataframe using time with date
dl_df = pd.DataFrame([
    dict(Machine=2, Start=0, Finish=4, job=1),
    dict(Machine=3, Start=0, Finish=3, job=2),
    dict(Machine=4, Start=0, Finish=2.6, job=3),
    dict(Machine=4, Start=2.6, Finish=3.8, job=4),
    dict(Machine=5, Start=0, Finish=2.5, job=5),
    dict(Machine=5, Start=2.5, Finish=4.3, job=6),
    dict(Machine=5, Start=4.3, Finish=7.7, job=7),
    dict(Machine=4, Start=3.8, Finish=5.5, job=8),
    dict(Machine=4, Start=5.5, Finish=7.7, job=9),
    dict(Machine=2, Start=4, Finish=5.5, job=10),
    dict(Machine=3, Start=3, Finish=7.4, job=11),
    dict(Machine=2, Start=5.5, Finish=7.5, job=12),
])
dl_df['delta'] = dl_df['Finish'] - dl_df['Start']

display(dl_df)
```

```python
colors = ['#FF0088',
          '#FF8800',
          '#FFFF00',
          '#77FF00',
          '#00FF99',
          '#00FFFF', 
          "#0066FF",
          "#7700FF",
          '#FF00FF',
          '#DDDDDD',
          '#FF0000',
          '#CCCCFF']


ylist = [0, (5, 4), (10, 4), (15, 4), (20, 4), (25, 4)]
```

```python
fig, ax = plt.subplots()


for i in range(dl_df.shape[0]):
    ax.broken_barh([(dl_df["Start"][i], dl_df["delta"][i])], ylist[dl_df["Machine"][i]], label=dl_df['job'][i], facecolors=colors[i])

ax.set_ylim(3, 30)
ax.set_xlim(0, 10)
ax.set_xlabel('Instance schedule')
ax.set_xticks(range(9))
ax.set_xticklabels(['7:30', '8:30', '9:30', '10:30', '11:30', '12:30', '13:30', '14:30', '15:30'])
ax.set_yticks([7,12,17,22,27])
ax.set_yticklabels(['1', '2', '3', '4', '5'])
ax.set_ylabel('Machine')
ax.grid(True)
ax.legend(loc="upper right")

#plt.show()
plt.savefig('fig1-1.png')
```
