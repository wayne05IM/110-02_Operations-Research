```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
plt.rcParams["figure.dpi"] = 150
```

```python
# create a dataframe using time with date
dl_df = pd.DataFrame([
    dict(Machine=2, Start=0.8, Finish=3.5, job=1, part=1),
    dict(Machine=5, Start=3.7, Finish=5.0, job=1, part=2),
    dict(Machine=3, Start=0.0, Finish=1.6, job=2, part=1),
    dict(Machine=3, Start=1.6, Finish=3.0, job=2, part=2),
    dict(Machine=5, Start=0.0, Finish=0.7, job=3, part=1),
    dict(Machine=4, Start=1.4, Finish=3.3, job=3, part=2),
    dict(Machine=5, Start=0.7, Finish=1.2, job=4, part=1),
    dict(Machine=4, Start=4.3, Finish=5.0, job=4, part=2),
    dict(Machine=2, Start=0.0, Finish=0.8, job=5, part=1),
    dict(Machine=4, Start=3.3, Finish=4.3, job=5, part=2),
    dict(Machine=2, Start=0.0, Finish=0.0, job=6, part=1),
    dict(Machine=5, Start=1.2, Finish=3.7, job=6, part=2),
    dict(Machine=4, Start=0.0, Finish=1.4, job=7, part=1),
    dict(Machine=3, Start=3.0, Finish=5.0, job=7, part=2),
    dict(Machine=3, Start=5.0, Finish=6.1, job=8, part=1),
    dict(Machine=4, Start=6.1, Finish=7.2, job=8, part=2),
    dict(Machine=4, Start=5.0, Finish=5.8, job=9, part=1),
    dict(Machine=3, Start=7.1, Finish=7.8, job=9, part=2),
    dict(Machine=3, Start=6.1, Finish=7.1, job=10, part=1),
    dict(Machine=5, Start=7.1, Finish=7.6, job=10, part=2),
    dict(Machine=2, Start=3.5, Finish=6.5, job=11, part=1),
    dict(Machine=2, Start=6.5, Finish=7.9, job=11, part=2),
    dict(Machine=5, Start=0.0, Finish=0.0, job=12, part=1),
    dict(Machine=5, Start=5.0, Finish=7.0, job=12, part=2),
])
dl_df['delta'] = dl_df['Finish'] - dl_df['Start']

display(dl_df)
```

```python
colors = [0,
          '#FF0088',
          '#FF8800',
          '#FFFF00',
          '#77FF00',
          '#00FF99',
          '#00FFFF', 
          "#0066FF",
          "#7700FF",
          '#FF00FF',
          '#DDDDDD',
          '#FF0000',
          '#CCCCFF']


ylist = [0, (5, 4), (10, 4), (15, 4), (20, 4), (25, 4)]
```

```python
fig, ax = plt.subplots()


for i in range(0, 24, 2):
    ax.broken_barh([(dl_df["Start"][i], dl_df["delta"][i])], ylist[dl_df["Machine"][i]], label=dl_df['job'][i], facecolors=colors[dl_df['job'][i]])
    ax.broken_barh([(dl_df["Start"][i+1], dl_df["delta"][i+1])], ylist[dl_df["Machine"][i+1]], facecolors=colors[dl_df['job'][i+1]])

ax.set_ylim(3, 30)
ax.set_xlim(0, 10)
ax.set_xlabel('Instance schedule')
ax.set_xticks(range(9))
ax.set_xticklabels(['7:30', '8:30', '9:30', '10:30', '11:30', '12:30', '13:30', '14:30', '15:30'])
ax.set_yticks([7,12,17,22,27])
ax.set_yticklabels(['1', '2', '3', '4', '5'])
ax.set_ylabel('Machine')
ax.grid(True)
ax.legend(loc="upper right")

plt.show()
```
